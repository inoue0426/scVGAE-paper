```python
import pandas as pd
import polars as pl
from scipy.io import mmread
```

```python
cluster = pd.read_csv('data/filtered_matrices_mex/hg19/analysis_csv/kmeans/10_clusters/clusters.csv')
cluster
```

```python
col = pd.read_csv('data/filtered_matrices_mex/hg19/barcodes.tsv', sep='\t', header=None)
col.columns = ['Barcode']
col
```

```python
col = col.merge(cluster, how='left')
col
```

```python
mtx = mmread('data/filtered_matrices_mex/hg19/matrix.mtx')
mtx = mtx.toarray()
```

```python
df = pd.DataFrame(mtx, columns=list(col.Cluster)).T
df
```

```python
df = df[~df.index.isna()]
df
```

```python
df.to_csv('data/pbmc40k/data.csv.gz', compression='gzip')
```
