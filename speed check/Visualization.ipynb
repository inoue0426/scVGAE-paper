```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = {
    'Baron': [4.15, 38.13, 66.78, 38.49, 33.38, 16.11],
    'Brosens': [6.06, 62.40, 52.83, 63.04, 50.34, 38.63],
    'Carey': [14.32, 131.48, 87.61, 63.60, 96.29, 103.89],
    'hcabm40k': [11.57, 38.41, 59.50, 97.34, 45.90, 34.49],
    'Chang': [6.45, 5.90, 19.52, 37.87, 12.47, 2.02]
}

index = ['MAGIC', 'ALRA', 'DeepImpute', 'DCA', 'GNNImpute', 'scVGAE']
df = pd.DataFrame(data, index=index)
```

```python
df.columns = [
    'Baron\n  (1937, 20125)', 'Brosens\n  (5829, 20697)', 'Carey\n  (9082, 27202)', 'hcabm40k\n  (12000, 17369)', 'Chang\n  (22770, 10160)'
]
```

```python
speed = pd.DataFrame()
for i in df.index:
    tmp = pd.DataFrame(df.loc[i].values, columns=['Value'])
    tmp['Dataset'] = [i]*len(df.columns)
    tmp['Methods'] =df.columns
    
    speed = pd.concat([
        speed, tmp
    ])
```

```python
cmap = sns.color_palette('colorblind')
fig = plt.figure(figsize=(10, 5))
sns.set(style='darkgrid')

ax = sns.barplot(data=speed[['Dataset', 'Methods', 'Value']], x='Methods', y='Value', hue='Dataset', palette=cmap)
ax.set(xlabel=None)
ax.set(ylabel='Speed (Sec)')
plt.show()
```
